```python
import pandas as pd #A 
import numpy as np #B 
from matplotlib import pyplot as plt #C 
import seaborn as sns #C 
%matplotlib inline
```

```python
df = pd.read_csv('sample_data/new_data3.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_') # A 
string_columns = list(df.dtypes[df.dtypes == 'object'].index) # B 
for col in string_columns: 
    df[col] = df[col].str.lower().str.replace(' ', '_') # C 
```

```python
df.columns
```

```python
plt.figure(figsize=(6, 4))

sns.histplot(df['ценна_доллар'], bins=40, color='black', alpha=1)
plt.ylabel('Frequency')
plt.xlabel('Price')
plt.title('Distribution of prices')

plt.show()
```

```python
log_price = np.log1p(df['ценна_доллар'])

plt.figure(figsize=(6, 4))

sns.histplot(log_price, bins=40, color='black', alpha=1)
plt.ylabel('Frequency')
plt.xlabel('Log(Price + 1)')
plt.title('Distribution of prices after log tranformation')

plt.show()
```

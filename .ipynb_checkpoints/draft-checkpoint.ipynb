```python
url = f'https://www.house.kg/kupit-kvartiru?page=1' 
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
Houses = soup.find_all('div', class_='main-wrapper')
```

```python
for each in Houses:
    link = each.find('div', class_='left-image').a['href']
    link = f'https://www.house.kg{link}'
    print(link)
    
    response1 = requests.get(link)
    soup1 = BeautifulSoup(response1.text, 'html.parser')
    
    series = soup1.find_all('div', class_='info-row')
   # print(series)
    house_SuggType = None
    house_Seria = None
    house_Dom = None
    house_Etaj = None
    house_Square = None
    house_Otop = None
    house_Sost = None
    house_Rasro = None
    house_Ipotek = None
    house_Obmen = None
    house_Internet = None
    house_Gas = None
    house_Balc = None
    house_Mebel = None
    house_Pol = None
    house_Potolok = None
    house_Besop = None
    house_docs = None
    house_tel = None
    house_sanuzel = None
    house_door = None
    house_parking = None
    
    for info_row in series:
        label = info_row.find('div', class_='label').text.strip()
        info = info_row.find('div', class_='info').text.strip()
        #print(f"{label}: {info}")
        
        if label == 'Тип предложения':
            house_SuggType = info
        elif label == 'Серия':
            house_Seria = info
        elif label == 'Дом':
            house_Dom = info
        elif label == 'Этаж':
            house_Etaj = info
        elif label == 'Площадь':
            house_Square = info
        elif label == 'Отопление':
            house_Otop = info
        elif label == 'Состояние':
            house_Sost = info
        elif label == 'Возможность рассрочки':
            house_Rasro = info
        elif label == 'Возможность ипотеки':
            house_Ipotek = info
        elif label == 'Возможность обмена':
            house_Obmen = info
        elif label == 'Интернет':
            house_Internet = info
        elif label == 'Газ':
            house_Gas = info
        elif label == 'Балкон':
            house_Balc = info
        elif label == 'Мебель':
            house_Mebel = info
        elif label == 'Пол':
            house_Pol = info
        elif label == 'Высота потолков':
            house_Potolok = info
        elif label == 'Безопасность':
            house_Besop = info
        elif label == 'Правоустанавливающие документы':
            house_docs = info
        elif label == 'Телефон':
            house_tel = info
        elif label == 'Санузел':
            house_sanuzel = info
        elif label == 'Входная дверь':
            house_door = info
        elif label == 'Парковка':
            house_parking = info


    priceDollar = soup1.find('div', class_='price-dollar').text.strip()
    priceSom = soup1.find('div', class_='price-som').text.strip()
    name = soup1.find('div', class_='details-header').find('h1').text.strip()
    print(name)
#   
    Namelist.append(name)
    PricelistDollar.append(priceDollar)
    PricelistSom.append(priceSom)
    SuggTypelist.append(house_SuggType)
    Serialist.append(house_Seria)
    Domlist.append(house_Dom)
    Etajlit.append(house_Etaj)
    Squarelist.append(house_Square)
    Otoplist.append(house_Otop)
    Sostlist.append(house_Sost)
    Rasrolist.append(house_Rasro)
    Ipoteklist.append(house_Ipotek)
    Obmenlist.append(house_Obmen)
    Internetlist.append(house_Internet)
    Gaslist.append(house_Gas)
    Balconylist.append(house_Balc)
    Mebellist.append(house_Mebel)
    Pollist.append(house_Pol)
    Potoloklist.append(house_Potolok)
    Besoplist.append(house_Besop)
    Docslist.append(house_docs)
    Tellist.append(house_tel)
    Sanuzellist.append(house_sanuzel)
    Doorlist.append(house_door)
    Parkinglist.append(house_parking)
```

```python

specifications = {'Тип предложения': SuggTypelist,
                          'Серия': Serialist,
                          'Дом': Domlist,
                          'Этаж': Etajlit,
                          'Площадь': Squarelist,
                  'Отопление': Otoplist,
                  'Состояние': Sostlist,
                  'Возможность рассрочки': Rasrolist,
                  'Возможность ипотеки':  Ipoteklist,
                  'Возможность обмена': Obmenlist,
                  'Интернет': Internetlist,
                  'Газ': Gaslist,
                  'Балкон': Balconylist,
                  'Мебель': Mebellist,
                  'Пол': Pollist,
                  'Высота потолков': Potoloklist,
                  'Безопасность': Besoplist,
                  'Правоустанавливающие документы': Docslist,
                  'Телефон': Tellist,
                  'Санузел': Sanuzellist,
                  'Входная дверь': Doorlist,
                  'Парковка': Parkinglist,
                  'Ценна доллар': PricelistDollar,
                  'Ценна сом': PricelistSom,
                  'Название': Namelist
                          }
df = pd.DataFrame(specifications)
df.to_csv('example4.csv')
```

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0]) # array with only ones
    X = np.column_stack([ones, X]) # add the ones array to the X (in the start of the every line)

    XTX = X.T.dot(X) # transpose of X and dot with X
    XTX_inv = np.linalg.inv(XTX) # inversion of XTX
    w = XTX_inv.dot(X.T).dot(y) # dot of inversion of the XTX and dot with y
    
    return w[0], w[1:] # split into bias and other w's
```

```python
base = ['площадь', 'высота_потолков', 'комнаты', 'номер_этажа', 'общий_этаж']
```

```python
def prepare_X(df):
    df_num = df[base] # use fetures only in base
    df_num = df_num.fillna(0) # fill missings with 0's
    X = df_num.values
    return X
```

```python
def rmse(y, y_pred):
    error = y_pred - y # differences between actual and the predicted data (numpy array)
    mse = (error ** 2).mean()  # squared and the mean
    return np.sqrt(mse) # square root (rmse)
```

```python
def prepare_X(df):
    df = df.copy()  # Create a copy of the input parameter to prevent side effects. 
    features = base.copy() # Create a copy of the base list with the basic features.

    features.append('охрана') # Append age to the list of feature names we use for the model. 

    features.append('красная_книга')
    features.append('технический_паспорт')
    features.append('решетки_на_окнах')
    
        
    df_num = df[features]
    df_num = df_num.fillna(0)
    X = df_num.values
    return X
```

```python
X_train = prepare_X(df_train)
w_0, w = train_linear_regression(X_train, y_train)

y_pred = w_0 + X_train.dot(w)
print('train:', rmse(y_train, y_pred))

X_val = prepare_X(df_val)
y_pred = w_0 + X_val.dot(w)
print('validation:', rmse(y_val, y_pred))
```

```python
def train_linear_regression_reg(X, y, r=0.0):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    reg = r * np.eye(XTX.shape[0]) # alpha with identity matrix
    XTX = XTX + reg

    XTX_inv = np.linalg.inv(XTX)
    w = XTX_inv.dot(X.T).dot(y)
    
    return w[0], w[1:]
```

```python
X_train = prepare_X(df_train)
```

```python
for r in [0, 0.001, 0.01, 0.1, 1, 10]:
    w_0, w = train_linear_regression_reg(X_train, y_train, r=r)
    print('%5s, %.2f, %.2f, %.2f' % (r, w_0, w[4], w[6]))
```

```python
X_train = prepare_X(df_train)
w_0, w = train_linear_regression_reg(X_train, y_train, r=0)

y_pred = w_0 + X_train.dot(w)
print('train', rmse(y_train, y_pred))

X_val = prepare_X(df_val)
y_pred = w_0 + X_val.dot(w)
print('val', rmse(y_val, y_pred))
```

```python
X_train = prepare_X(df_train)
w_0, w = train_linear_regression_reg(X_train, y_train, r=0.01)

y_pred = w_0 + X_train.dot(w)
print('train', rmse(y_train, y_pred))

X_val = prepare_X(df_val)
y_pred = w_0 + X_val.dot(w)
print('val', rmse(y_val, y_pred))
```

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0]) # array with only ones
    X = np.column_stack([ones, X]) # add the ones array to the X (in the start of the every line)

    XTX = X.T.dot(X) # transpose of X and dot with X
    XTX_inv = np.linalg.inv(XTX) # inversion of XTX
    w = XTX_inv.dot(X.T).dot(y) # dot of inversion of the XTX and dot with y
    
    return w[0], w[1:] # split into bias and other w's
```

```python
base = ['площадь', 'высота_потолков', 'комнаты', 'номер_этажа', 'охрана', 'общий_этаж']
```

```python
def prepare_X(df):
    df_num = df[base] # use fetures only in base
    #df_num = df_num.fillna(0) # fill missings with 0's
    X = df_num.values
    return X
```

```python
def rmse(y, y_pred):
    error = y_pred - y # differences between actual and the predicted data (numpy array)
    mse = (error ** 2).mean()  # squared and the mean
    return np.sqrt(mse) # square root (rmse)
```

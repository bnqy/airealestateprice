```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
import re
```

```python
def dataColl(pagenum):
    try:
        linklist = []  
        FromCenterlist = []
        SquareOveralllist = []
        Repairlist = []
        YearofConstructionlist = []
        FloorNumberlist = []
        Walllist = []
        KitchenArealist = []
        RoomNumberlist = []
        CeilingHeightlist = []
        Toiletlist = []
        Balconylist = []
        WindowMateriallist = []
        SunSidelist = []
        ViewFromWindowlist = []
        Liftlist = []
        Ramplist = []
        Yardlist = []
        Parkinglist = []
        Pricelist = []
        Addresslist = []
    
        for num in range(1, pagenum):
            url = f'https://bishkek.etagi.com/realty/?page={num}' 
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')
            Phones = soup.find_all('div', class_='templates-object-card')

            for each in Phones:
                #link = each.find('div', class_='templates-object-card__slider').a['href']  # First I had to find the links for individual product pages in the search result page.
                #link = f'https://bishkek.etagi.com/{link}'
                #linklist.append(link)
                #print(link)
                link_element = each.find('a')
                link = f'https://bishkek.etagi.com{link_element["href"]}'
                linklist.append(link)

                response1 = requests.get(link)  # Then I created another soup for product-specific data.
                soup1 = BeautifulSoup(response1.text, 'html.parser')
                data = soup1.find('ul', class_='PpfZ1').text


                Batterymatch = re.search('Общая площадь', data)
                if Batterymatch:
                    BatteryPower = Batterymatch.group(1)
                else:
                    BatteryPower = None
                SquareOveralllist.append(BatteryPower)
            
            
        specifications = {'Model': SquareOveralllist}
        df = pd.DataFrame(specifications)
        df.to_csv('Trendyol_detailed333.csv')
        
    except ValueError as e:
        print(e)
```

```python
for each in Phones:
    link = each.find('div', class_='left-image').a['href']  # First I had to find the links for individual product pages in the search result page.
    link = f'https://www.house.kg{link}'
    print(link)
    
    response1 = requests.get(link)  # Then I created another soup for product-specific data.
    soup1 = BeautifulSoup(response1.text, 'html.parser')
    # data = soup1.find('ul', class_='detail-attr-container').text
    
    
    series = soup1.find_all('div', class_='info-row')
    for info_row in series:
        label = info_row.find('div', class_='label').text.strip()
        info = info_row.find('div', class_='info').text.strip()
        print(f"{label}: {info}")
        
        if label == 'Тип предложения':
            SuggTypelist.append(info)
        else: 
            SuggTypelist.append(None)
    
        if label == 'Серия':
            Serialist.append(info)
        else: 
            Serialist.append(None)
        
        if label == 'Дом':
            Domlist.append(info)
        else: 
            Domlist.append(None)
            
        if label == 'Этаж':
            Etajlit.append(info)
        else: 
            Etajlit.append(None)
        
        if label == 'Площадь':
            Squarelist.append(info)
        else: 
            Squarelist.append(None)
```

```python
specifications = {'Тип предложения': SuggTypelist,
                          'Серия': Serialist,
                          'Дом': Domlist,
                          'Этаж': Etajlit,
                          'Площадь': Squarelist
                          }
```

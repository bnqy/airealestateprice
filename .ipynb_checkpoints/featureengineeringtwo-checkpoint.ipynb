```python
import pandas as pd
import csv
import numpy as np 
import seaborn as sns 
from matplotlib import pyplot as plt 
%matplotlib inline 

df = pd.read_csv('collect_lon_lat.csv')
df.head()
```

```python
len(df)
```

```python
df['Город'] = df['Адрес'].str.split(',').str[0]
df['Город'] = df['Город'].str.replace(' Ленинский район', '')
df['Город'] = df['Город'].str.replace(' Октябрьский район', '')
df['Город'] = df['Город'].str.replace(' Свердловский район', '')
```

```python
df = df[df['Город'] == 'Бишкек']
df.groupby('Город')['Город'].agg('count')
```

```python
df['Площадь'] = df['Площадь'].str.extract(r'^(\d+\.?\d*)')
df['Ценна доллар'] = df['Ценна доллар'].str.replace('$', '')
df['Ценна доллар'] = df['Ценна доллар'].str.replace(' ', '')
df['Высота потолков'] = df['Высота потолков'].str.replace('м.', '')
df['Название'] = df['Название'].str.extract(r'^(\d)')
df[['Номер этажа', 'Общий этаж']] = df['Этаж'].str.split(' этаж из ', expand=True)
df = df.rename(columns={'Название': 'Комнаты'})
```

```python
df = df.dropna(subset=['Общий этаж'])
del df['Ценна сом']
```

```python
df['Ценна доллар'] = pd.to_numeric(df['Ценна доллар'], errors='coerce')
df['Площадь'] = pd.to_numeric(df['Площадь'], errors='coerce')
df['Комнаты'] = pd.to_numeric(df['Комнаты'], errors='coerce')
df['Номер этажа'] = pd.to_numeric(df['Номер этажа'], errors='coerce')
df['Общий этаж'] = pd.to_numeric(df['Общий этаж'], errors='coerce')
df['Высота потолков'] = pd.to_numeric(df['Высота потолков'], errors='coerce')
```

```python
df['Правоустанавливающие документы'] = df['Правоустанавливающие документы'].str.replace(r'\s+', ' ', regex=True)
df['Безопасность'] = df['Безопасность'].str.replace(r'\s+', ' ', regex=True)
```

```python
df['Домофон'] = df['Безопасность'].str.contains('домофон', case=False, na=False).astype(int)

df['Видеонаблюдение'] = df['Безопасность'].str.contains('видеонаблюдение', case=False, na=False).astype(int)

df['Охрана'] = df['Безопасность'].str.contains('охрана', case=False, na=False).astype(int)

df['Решетки на окнах'] = df['Безопасность'].str.contains('решетки на окнах', case=False, na=False).astype(int)

df['Кодовый замок'] = df['Безопасность'].str.contains('кодовый замок', case=False, na=False).astype(int)

df['Сигнализация'] = df['Безопасность'].str.contains('сигнализация', case=False, na=False).astype(int)

df['Видеодомофон'] = df['Безопасность'].str.contains('видеодомофон', case=False, na=False).astype(int)

df['Консьерж'] = df['Безопасность'].str.contains('консьерж', case=False, na=False).astype(int)
```

```python
df['Договор купли-продажи'] = df['Правоустанавливающие документы'].str.contains('договор купли-продажи', case=False, na=False).astype(int)

df['Договор дарения'] = df['Правоустанавливающие документы'].str.contains('договор дарения', case=False, na=False).astype(int)

df['Технический паспорт'] = df['Правоустанавливающие документы'].str.contains('технический паспорт', case=False, na=False).astype(int)

df['Акт ввода в эксплуатацию'] = df['Правоустанавливающие документы'].str.contains('акт ввода в эксплуатацию', case=False, na=False).astype(int)

df['Красная книга'] = df['Правоустанавливающие документы'].str.contains('красная книга', case=False, na=False).astype(int)

df['зеленая книга'] = df['Правоустанавливающие документы'].str.contains('зеленая книга', case=False, na=False).astype(int)
```

```python
del df['Правоустанавливающие документы']
del df['Безопасность']
del df['Этаж']
```

```python
series_mapping = {
    '106 серия улучшенная': '106 серия',
    '105 серия улучшенная': '105 серия',
    '104 серия улучшенная': '104 серия'
}

df['Серия'] = df['Серия'].replace(series_mapping)
```

```python
df[['Материал', 'Год']] = df['Дом'].str.split(',', expand=True)

df['Материал'] = df['Материал'].str.strip()
df['Год'] = df['Год'].str.strip()
```

```python
del df['Дом']
```

```python
df['Год'] = df['Год'].str.replace(' г.', '')
df.groupby('Год')['Год'].agg('count').sort_values(ascending=False)
```

```python
def transform_value(x):
    if x > 50:
        return x / 100
    else:
        return x
    
df['Высота потолков'] = df['Высота потолков'].apply(transform_value)

def transform_value2(x):
    if x > 10:
        return x / 10
    else:
        return x
    
df['Высота потолков'] = df['Высота потолков'].apply(transform_value2)
```

```python
df.groupby('Высота потолков')['Высота потолков'].agg('count').sort_values(ascending=False)
```

```python
df["Стоимость, доллар США за 1 кв.м"] = df["Ценна доллар"]/df["Площадь"]
```

```python
df["Тип предложения"] = df["Тип предложения"].replace({"от собственника": "собственник", "от агента": "агент"})
```

```python
df.head().T
```

```python
otop_mapping = {
    'на газе': 'есть',
    'электрическое': 'есть',
    'центральное': 'есть',
    'смешанное': 'есть',
    'автономное': 'есть',
    'без отопления': 'нет'
}

df['Отопление'] = df['Отопление'].replace(otop_mapping)
```

```python
df.isnull().sum()
```

```python
gaz_mapping = {
    'автономный': 'есть',
    'магистральный': 'есть',
    'возможно подключение': 'нет'
}

df['Газ'] = df['Газ'].replace(gaz_mapping)
```

```python

balkon_mapping = {
    'застекленный балкон': 'есть',
    'лоджия': 'есть',
    'балкон': 'есть'
}

df['Балкон'] = df['Балкон'].replace(balkon_mapping)
```

```python


tel_mapping = {
    'возможно подключение': 'нет'
}

df['Телефон'] = df['Телефон'].replace(tel_mapping)
```

```python
df['Отопление'] = df['Отопление'].replace({'нет': 0, 'есть': 1})
df['Отопление'] = pd.to_numeric(df['Отопление'], errors='coerce')

df['Возможность рассрочки'] = df['Возможность рассрочки'].replace({'нет': 0, 'есть': 1})
df['Возможность рассрочки'] = pd.to_numeric(df['Возможность рассрочки'], errors='coerce')

df['Возможность ипотеки'] = df['Возможность ипотеки'].replace({'нет': 0, 'есть': 1})
df['Возможность ипотеки'] = pd.to_numeric(df['Возможность ипотеки'], errors='coerce')

df['Газ'] = df['Газ'].replace({'нет': 0, 'есть': 1})
df['Газ'] = pd.to_numeric(df['Газ'], errors='coerce')

df['Балкон'] = df['Балкон'].replace({'нет': 0, 'есть': 1})
df['Балкон'] = pd.to_numeric(df['Балкон'], errors='coerce')


df['Телефон'] = df['Телефон'].replace({'нет': 0, 'есть': 1})
df['Телефон'] = pd.to_numeric(df['Телефон'], errors='coerce')
```

```python
column_names = df.columns.tolist()

def print_distinct_values(df, columns):
    for column in columns:
        unique_values = df[column].unique()
        print(f"Distinct values for column '{column}':")
        print(unique_values)
        print()

print_distinct_values(df, column_names)
```

```python

numeric_cols = df.select_dtypes(include=['number']).columns
for i in range(len(numeric_cols)):
    df[numeric_cols[i]].fillna(0, inplace=True)
    
if all(col in df.columns for col in numeric_cols):
    daf = df[numeric_cols]
    corr_matrix = daf.corr()

    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('Корреляционная матрица')
    plt.show()
else:
    print("Неправильные данные")
```
